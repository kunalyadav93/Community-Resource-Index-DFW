```python
import numpy as np
import pandas as pd
```

### 1. Employment Rate

Table Name : B23001

URL : https://data.census.gov/cedsci/table?q=B23001%3A%20SEX%20BY%20AGE%20BY%20EMPLOYMENT%20STATUS%20FOR%20THE%20POPULATION%2016%20YEARS%20AND%20OVER&tid=ACSDT5Y2017.B23001&hidePreview=true&tp=true&vintage=2017&g=0500000US48085.140000,48113.140000,48121.140000,48439.140000&t=Employment%20and%20Labor%20Force%20Status&moe=false

```python
employment= pd.read_csv("ACSDT5Y2018.B23001_data_with_overlays_2020-04-14T132501.csv",header=1)
```

#### Removing error columns

```python
cols=[i for i in employment.columns if "Error" not in i]
employment=employment[cols]
employment.head(5)
```

#### Creating the column Employment Rate

```python
employment['In_labor_force(population)']=employment.loc[:,employment.columns.str.endswith("In labor force")].sum(axis=1)
employment["Employment_rate(%)(population)"]=employment['In_labor_force(population)']/employment['Estimate!!Total']*100
```

```python
employment=employment[['id','Geographic Area Name','Estimate!!Total',
            'In_labor_force(population)','Employment_rate(%)(population)'
           ]].rename(columns={'Estimate!!Total':'Total_Population'})
```

```python
employment
```

### 2. Commute Duration
Table : B08303

```python
commute_df=pd.read_csv("ACSDT5Y2018.B08303_data_with_overlays_2020-04-14T130800.csv",header=1)
```

#### Removing error columns

```python
cols=[i for i in commute_df.columns if "Error" not in i]
commute_df=commute_df[cols]
commute_df.head(5)
```

#### Creating the column Commute Duration 60+ minutes

```python
commute_df["Commute_Duration_60+(pcw)"]=commute_df['Estimate!!Total!!60 to 89 minutes']+commute_df['Estimate!!Total!!90 or more minutes']
commute_df['% Population_Work_Commute_Duration_60+(pcw)']=commute_df["Commute_Duration_60+(pcw)"]/commute_df['Estimate!!Total']*100
```

#### Creating the column Commute Duration 90+ minutes

```python
commute_df['% Population_Work_Commute_Duration_90+(pcw)']=commute_df["Estimate!!Total!!90 or more minutes"]/commute_df['Estimate!!Total']*100
```

```python
commute_df=commute_df[['id','Geographic Area Name','Estimate!!Total','Commute_Duration_60+(pcw)',
            '% Population_Work_Commute_Duration_60+(pcw)','Estimate!!Total!!90 or more minutes',
            '% Population_Work_Commute_Duration_90+(pcw)'
           ]].rename(columns={'Estimate!!Total':'Population_commuting_to_work(pcw)',
                              'Estimate!!Total!!90 or more minutes':'Commute_Duration_90+(pcw)'})
```

```python
commute_df
```

### 3. Family below poverty level
Table: B17010

```python
fam_df=pd.read_csv("ACSDT5Y2018.B17010_data_with_overlays_2020-04-14T132004.csv",header=1)
```

#### Removing error columns

```python
cols=[i for i in fam_df.columns if "Error" not in i]
fam_df=fam_df[cols]
fam_df.head(5)
```

#### Creating column : Families below poverty level

```python
fam_df['% family_below_poverty_level(family)']=fam_df['Estimate!!Total!!Income in the past 12 months below poverty level']/fam_df['Estimate!!Total']*100
```

#### Creating column : Single-Headed families below poverty level

```python
fam_df["% Single_headed_family_below_poverty_level(family)"]=fam_df['Estimate!!Total!!Income in the past 12 months below poverty level!!Other family']/fam_df['Estimate!!Total']*100
```

```python
fam_df=fam_df[['id','Geographic Area Name','Estimate!!Total',
        '% family_below_poverty_level(family)',
        '% Single_headed_family_below_poverty_level(family)'
       ]].rename(columns={'Estimate!!Total':'Total_family'})
```

```python
fam_df
```

### 6. House ownership rate
Table DP04

```python
house_ownership=pd.read_csv("ACSDP5Y2018.DP04_data_with_overlays_2020-04-19T054110.csv",header=1)
```

#### Removing error column

```python
cols=[i for i in house_ownership.columns if "Error" not in i]
house_ownership=house_ownership[cols]
house_ownership.head(5)
```

```python
house_ownership[["id","Geographic Area Name",
                 "Estimate!!VALUE!!Owner-occupied units",
                 "Estimate!!HOUSING OCCUPANCY!!Total housing units"]]
```

#### Creating column: House ownership rate

```python
house_ownership["% house_ownership(housing_units)"]=house_ownership["Estimate!!VALUE!!Owner-occupied units"]/house_ownership["Estimate!!HOUSING OCCUPANCY!!Total housing units"]*100
```

```python
house_ownership=house_ownership[["id","Geographic Area Name","% house_ownership(housing_units)",
                 "Estimate!!HOUSING OCCUPANCY!!Total housing units"
                ]].rename(columns={"Estimate!!HOUSING OCCUPANCY!!Total housing units":"Total_housing_units"})
```

```python
house_ownership
```

### 7. Public Assistance
Table B19057

```python
public_assistance=pd.read_csv("ACSDT5Y2018.B19057_data_with_overlays_2020-04-19T062106.csv",header=1)
```

```python
public_assistance=public_assistance[["id","Geographic Area Name","Estimate!!Total!!With public assistance income","Estimate!!Total"
                  ]].rename(columns={"Estimate!!Total!!With public assistance income":"households_with_public_assistance_income",
                                    "Estimate!!Total":"total_households"})
```

```python
public_assistance
```

### 8. SNAP
Table S2201

```python
snap=pd.read_csv("ACSST5Y2018.S2201_data_with_overlays_2020-04-19T062707.csv",header=1)
```

```python
snap=snap[["id","Geographic Area Name","Estimate!!Households receiving food stamps/SNAP!!Households"
     ]].rename(columns={"Estimate!!Households receiving food stamps/SNAP!!Households":"households_recieving_SNAP"})
```

### Joining all the datasets 

```python
df_1=employment.merge(commute_df,on=["id","Geographic Area Name"],how='inner')
```

```python
df_2=df_1.merge(fam_df,on=["id","Geographic Area Name"],how='inner') 
```

```python
df_3=df_2.merge(median_household,on=["id","Geographic Area Name"],how='inner')
```

```python
df_4=df_3.merge(gini,on=["id","Geographic Area Name"],how='inner')
```

```python
df_5=df_4.merge(house_ownership,on=["id","Geographic Area Name"],how='inner')
```

```python
df_6=df_5.merge(public_assistance,on=["id","Geographic Area Name"],how='inner')
```

```python
df_7=df_6.merge(snap,on=["id","Geographic Area Name"],how='inner')
```

### Final Dataset containing all columns

```python
df_7.to_csv("Economic_Indicators_All_Counties.csv")
```
